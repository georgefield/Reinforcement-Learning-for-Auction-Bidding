```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def custom_function(epsilon):
    return 10 * (0.5 + epsilon)**2 * (0.5 - epsilon)**3 + 5 * (0.5 + epsilon) * (0.5 - epsilon)**4 + (0.5 - epsilon)**5

# Generate values for epsilon from 0 to 0.5
epsilon_values = np.linspace(0, 0.5, 100)
function_values = custom_function(epsilon_values)

# Plot the function
plt.figure(figsize=(5, 1.5))
plt.plot(epsilon_values, function_values, label='')
plt.title('Plot of the Function')
plt.xlabel('Epsilon')
plt.ylabel('Function Value')
plt.grid(True)
plt.legend()
plt.show()
```
